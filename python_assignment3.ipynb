## Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()

```python
def addition(a,b):
    return a+b
def myreduce(self,l):
    a=l[0]
    for i in range(1,len(l)):
        b=l[i]
        a=self(a,b)
    return a
        
    
    
        
```

```python
myreduce(addition,[1,2,3,4])
```

## Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()

```python
def myfilter(self,l):
    list1=[]
    for i in l:
        if(self(i)):
            list1.append(i)
    return list1 
```

```python
def is_mul5(a):
    if(a%5==0):
        return True
    return False
myfilter(is_mul5,[1,3,45,43,66,67,50,45])
```
