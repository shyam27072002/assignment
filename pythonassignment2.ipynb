## Create the below pattern using nested for loop in Python. 
* 
* * 
* * * 
* * * * 
* * * * * 
* * * * 
* * * 
* * 
* 


```python
n=1
i=1
while(n>0):
    print('*'*n)
    i+=1
    if(i>5):
        n-=1
    else:
        n+=1
```

## Write a Python program to reverse a word after accepting the input from the user. Sample Output: 
Input word: ineuron 
Output: norueni


```python
inp_word = input("enter word : ")
print("Input word  : ",inp_word)
print("Output : ",inp_word[::-1])
```
